```python
import time
import requests
import json

API_URL = "http://localhost:4000/api/v1/search/new"
DELAY_SECONDS = 9  # delay between queries

# ============================
#  VALID QUESTIONS FROM CSV
# ============================
queries = [

    # -------- Basic Feature Retrieval (20)
    "What are the features of DDoS-RSTFINFlood?",
    "What is the weight of DDoS-RSTFINFlood?",
    "What features are associated with DoS-TCP_Flood?",
    "What are the features of DDoS-ICMP_Flood?",
    "Show me the features of DDoS-SynonymousIP_Flood.",
    "What are the features of Mirai-udpplain?",
    "What are the features of Mirai-greip_flood?",
    "What features describe DDoS-SYN_Flood?",
    "What are the features of DDoS-PSHACK_Flood?",
    "What are the features of DDoS-TCP_Flood?",
    "What are the features of DDoS-UDP_Flood?",
    "What features describe MITM-ArpSpoofing?",
    "What are the features of DDoS-ACK_Fragmentation?",
    "What numeric features describe DoS-HTTP_Flood?",
    "What are the features of DDoS-ICMP_Fragmentation?",
    "What features describe Recon-PortScan?",
    "What are the features of DNS_Spoofing?",
    "What features describe Recon-OSScan?",
    "What are the features of Recon-HostDiscovery?",
    "What are the features of XSS?",

    # -------- Weight Questions (5)
    "What is the weight of DoS-TCP_Flood?",
    "What is the weight of DDoS-ICMP_Flood?",
    "What is the weight of DDoS-UDP_Fragmentation?",
    "What is the weight of VulnerabilityScan?",
    "What is the weight of BrowserHijacking?",

    # -------- Operations / Reasoning based (15)
    "Compare the features of DDoS-SYN_Flood and DoS-SYN_Flood.",
    "Which attack has a higher average flow duration: DDoS-UDP_Flood or DDoS-TCP_Flood?",
    "List the top 5 numeric features that change the most for DDoS-ICMP_Flood.",
    "Summarize the differences between Mirai-udpplain and Mirai-greeth_flood.",
    "Which protocol number appears most frequently for Recon-PortScan?",
    "Which label has the smallest header length value in the dataset?",
    "Give a statistical summary for DDoS-TCP_Flood (min, max, mean if available).",
    "What is the expected packet behavior pattern of DDoS-PSHACK_Flood?",
    "Which feature varies the most for MITM-ArpSpoofing?",
    "Does DDoS-ICMP_Flood have mostly zero values? Explain.",
    "Compare BenignTraffic with Recon-HostDiscovery.",
    "Which attack shows the highest number of zero-valued features?",
    "What feature uniquely identifies DNS_Spoofing based on the dataset?",
    "What is a distinguishing characteristic between Backdoor_Malware and BrowserHijacking?",
    "Give a high-level explanation of how to detect XSS using the dataset features.",

    # -------- Additional Attacks (10)
    "What are the features of DDoS-UDP_Fragmentation?",
    "What are the features of CommandInjection?",
    "What are the features of VulnerabilityScan?",
    "What are the features of DDoS-SlowLoris?",
    "What features describe Backdoor_Malware?",
    "What are the features of BrowserHijacking?",
    "What are the features of BenignTraffic?",
    "Which features show anomalies in Mirai-greeth_flood?",
    "Which flow properties are extreme in DDoS-ICMP_Fragmentation?",
    "Explain the behavior of DoS-UDP_Flood using its feature distribution."

]

print(f"Starting accuracy test for {len(queries)} queries...\n")

results = []

# ---------------------------
# Retry-safe request function
# ---------------------------
def send_query_with_retry(payload, retries=3):
    for attempt in range(1, retries + 1):
        try:
            return requests.post(API_URL, json=payload, timeout=60)
        except Exception as e:
            print(f"⚠️ Attempt {attempt}/{retries} failed: {e}")
            if attempt < retries:
                time.sleep(5)
            else:
                return None


# ---------------------------
# Main Test Loop
# ---------------------------
for idx, query in enumerate(queries, 1):
    print(f"[{idx}/{len(queries)}] Sending query: {query}")

    start_time = time.time()
    response = send_query_with_retry({"query": query})
    elapsed = round(time.time() - start_time, 2)

    if response is None:
        answer = "EXCEPTION: Failed after retries"
        print("✗ No response from server.\n")
    else:
        if response.status_code == 200:
            try:
                answer = response.json().get("answer", "").strip()
            except Exception:
                answer = f"BAD JSON RESPONSE: {response.text}"
        else:
            answer = f"ERROR {response.status_code}: {response.text}"

        print(f"✓ Received in {elapsed}s")
        print(f"Answer: {answer[:150]}...\n")

    results.append({
        "query": query,
        "answer": answer,
        "response_time_sec": elapsed
    })

    print(f"⏳ Waiting {DELAY_SECONDS} seconds before next query...\n")
    time.sleep(DELAY_SECONDS)

# ---------------------------
# Save results to JSON
# ---------------------------
with open("accuracy_results.json", "w") as f:
    json.dump(results, f, indent=4)

print("🎉 Done! Results saved to accuracy_results.json")
```
