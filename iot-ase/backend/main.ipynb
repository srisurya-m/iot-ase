```python
import pandas as pd

# Path to your input CSV file
input_file = 'IoT_Intrusion.csv'

# Path to save the output CSV file
output_file = 'iot_intrusion_reduced.csv'

# Read the CSV file
df = pd.read_csv(input_file)

# Select the first 5001 rows
df_first_5001 = df.head(5001)

# Save the selected rows to a new CSV file
df_first_5001.to_csv(output_file, index=False)

print(f"First 5001 rows saved to {output_file}")
```
